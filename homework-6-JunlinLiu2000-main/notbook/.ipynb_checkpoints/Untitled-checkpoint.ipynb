```python
import numpy as np
import pandas as pd

data = pd.read_csv('../Data/wdbc.data', header=None)
data

data.replace('M', 1, inplace=True)
data.replace('B', 0, inplace=True)

pos_data = data[data[1]==1]
neg_data = data[data[1]==0]
pos_data
neg_data
```

## 1(b)i

```python
from sklearn.model_selection import train_test_split

x_train_pos, x_test_pos, y_train_pos, y_test_pos = train_test_split(pos_data.iloc[:,2:], pos_data.iloc[:,1], test_size=0.2, random_state=42)
x_train_neg, x_test_neg, y_train_neg, y_test_neg = train_test_split(neg_data.iloc[:,2:], neg_data.iloc[:,1], test_size=0.2, random_state=42)
x_train_neg

x_train = x_train_pos.append(x_train_neg)
x_test = x_test_pos.append(x_test_neg)
y_train = y_train_pos.append(y_train_neg)
y_test = y_test_pos.append(y_test_neg)
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

scaler = MinMaxScaler()
stand_x_train = scaler.fit_transform(x_train)
stand_x_test = scaler.fit_transform(x_test)

c_list = []
for i in range(-3,3):
    c_list.append(10**i)
parameters = {'C':c_list}

train_accuracy_list = []
train_precision_list = []
train_recall_list = []
train_F1_score_list = []
train_AUC_list=[]

test_accuracy_list = []
test_precision_list = []
test_recall_list = []
test_F1_score_list = []
test_AUC_list=[]

for i in range(30):
    lin_svc = LinearSVC(penalty='l1', dual=False, random_state=i)
    lin_clf_train = GridSearchCV(lin_svc, parameters, cv=5).fit(stand_x_train, y_train)
    lin_C_train = lin_clf_train.best_params_['C']
    lin_C_train
    
    lin_clf_test = GridSearchCV(lin_svc, parameters, cv=5).fit(stand_x_test, y_test)
    lin_C_test = lin_clf_test.best_params_['C']
    lin_C_test
    
    svm_train = LinearSVC(penalty='l1', C = lin_C_train, dual=False, random_state=1).fit(stand_x_train, y_train)
    y_train_predict = svm_train.predict(stand_x_train)

    svm_test = LinearSVC(penalty='l1', C = lin_C_test, dual=False, random_state=1).fit(stand_x_test, y_test)
    y_test_predict = svm_test.predict(stand_x_test)
    
    train_accuracy = accuracy_score(y_train, y_train_predict)
    train_accuracy_list.append(train_accuracy)
    test_accuracy = accuracy_score(y_test, y_test_predict)
    test_accuracy_list.append(test_accuracy)
    
    train_confusion_matrix = confusion_matrix(y_train, y_train_predict)
    test_confusion_matrix = confusion_matrix(y_test, y_test_predict)
    
    TN_train = train_confusion_matrix[0,0]
    FP_train = train_confusion_matrix[0,1]
    FN_train = train_confusion_matrix[1,0]
    TP_train = train_confusion_matrix[1,1]
    train_recall = (TP_train/(TP_train+FN_train))
    train_precision = (TP_train/(TP_train+FP_train))
    train_f1_score = 2*train_precision*train_recall/(train_precision+train_recall)
    
    train_precision_list.append(train_precision)
    train_recall_list.append(train_recall)
    train_F1_score_list.append(train_f1_score)
    
    TN_test = test_confusion_matrix[0,0]
    FP_test = test_confusion_matrix[0,1]
    FN_test = test_confusion_matrix[1,0]
    TP_test = test_confusion_matrix[1,1]
    test_recall = (TP_test/(TP_test+FN_test))
    test_precision = (TP_test/(TP_test+FP_test))
    test_f1_score = 2*test_precision*test_recall/(test_precision+test_recall)
    
    test_precision_list.append(test_precision)
    test_recall_list.append(test_recall)
    test_F1_score_list.append(test_f1_score)
    
    train_AUC = roc_auc_score(y_train, y_train_predict)
    test_AUC = roc_auc_score(y_test, y_test_predict)
    train_AUC_list.append(train_AUC)
    test_AUC_list.append(test_AUC)
```

```python
average_train_accuracy = np.mean(train_accuracy_list)
average_train_precision = np.mean(train_precision_list)
average_train_recall = np.mean(train_recall_list)
average_train_F1_score = np.mean(train_F1_score_list)
average_train_AUC = np.mean(train_AUC_list)
print(f"average_train_accuracy is:", average_train_accuracy)
print(f"average_train_precision is:", average_train_precision)
print(f"average_train_recall is:", average_train_recall)
print(f"average_train_F1_score is:", average_train_F1_score)
print(f"average_train_AUC is:", average_train_AUC)

average_test_accuracy = np.mean(test_accuracy_list)
average_test_precision = np.mean(test_precision_list)
average_test_recall = np.mean(test_recall_list)
average_test_F1_score = np.mean(test_F1_score_list)
average_test_AUC = np.mean(test_AUC_list)
print(f"\naverage_test_accuracy is:", average_test_accuracy)
print(f"average_test_precision is:", average_test_precision)
print(f"average_test_recall is:", average_test_recall)
print(f"average_test_F1_score is:", average_test_F1_score)
print(f"average_test_AUC is:", average_test_AUC)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
train_confusion_matrix = confusion_matrix(y_train, y_train_predict)
print("train_confusion_matrix is: \n", train_confusion_matrix)

fpr = dict()
tpr = dict()
roc_auc = dict()
fpr, tpr, _ = roc_curve(y_train, y_train_predict)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('train roc curve')
plt.legend(loc="lower right")
plt.show()
```

## (b)ii B

```python
lab_train_accuracy_list = []
lab_train_precision_list = []
lab_train_recall_list = []
lab_train_F1_score_list = []
lab_train_AUC_list=[]

lab_test_accuracy_list = []
lab_test_precision_list = []
lab_test_recall_list = []
lab_test_F1_score_list = []
lab_test_AUC_list=[]

for i in range(30):
    lab_svc = LinearSVC(penalty='l1', dual=False, random_state = i)
    lab_clf_train = GridSearchCV(lab_svc, parameters, cv=5).fit(labeled_x_train, labeled_y_train)
    lab_C_train = lab_clf_train.best_params_['C']
    
    svm_label_train = LinearSVC(penalty='l1', C = lab_C_train, dual=False, random_state = i)
    svm_label_train.fit(labeled_x_train, labeled_y_train)
    y_train_predict = svm_label_train.predict(labeled_x_train)
    
    unlabeled_x_train1 = unlabeled_x_train
    while len(unlabeled_x_train1) > 0:
        decision_boundary = svm_label_train.decision_function(unlabeled_x_train1)
        distance = np.abs(decision_boundary)
        max_dis = np.max(distance)
        for i in range(len(distance)):
            if distance[i] == max_dis:
                index_max = i
                #print(f"farthest distance is:", max_dis)
        if decision_boundary[index_max] < 0:
            pred_label = 0
        else:
            pred_label = 1
            #print(index_max)

        x = pd.DataFrame(np.array(unlabeled_x_train1.iloc[index_max]).reshape(1,len(unlabeled_x_train1.iloc[index_max])))
        new_labeled_x_train = labeled_x_train.append(x)
        new_labeled_y_train = pd.DataFrame(labeled_y_train.append(pd.Series(pred_label)))
        unlabeled_x_train1 = unlabeled_x_train1.drop(unlabeled_x_train1.index[index_max])
        new_lab_svc = LinearSVC(penalty='l1', dual=False, random_state = i)
        new_lab_clf_train = GridSearchCV(new_lab_svc, parameters, cv=5).fit(new_labeled_x_train, new_labeled_y_train)
        new_lab_C_train = new_lab_clf_train.best_params_['C']
        svm_label_train1 = LinearSVC(penalty='l1', C = new_lab_C_train, dual=False, random_state = i)
        svm_label_train1.fit(new_labeled_x_train, new_labeled_y_train)
    
    lab_y_train_predict = svm_label_train.predict(new_labeled_x_train)
    lab_y_test_predict = svm_label_train.predict(pd.DataFrame(x_test))
        
    lab_train_accuracy = accuracy_score(new_labeled_y_train, lab_y_train_predict)
    lab_train_accuracy_list.append(lab_train_accuracy)
    lab_test_accuracy = accuracy_score(y_test, lab_y_test_predict)
    lab_test_accuracy_list.append(lab_test_accuracy)
    
    lab_train_confusion_matrix = confusion_matrix(new_labeled_y_train, lab_y_train_predict)
    lab_test_confusion_matrix = confusion_matrix(y_test, lab_y_test_predict)
    
    TN_train = lab_train_confusion_matrix[0,0]
    FP_train = lab_train_confusion_matrix[0,1]
    FN_train = lab_train_confusion_matrix[1,0]
    TP_train = lab_train_confusion_matrix[1,1]
    train_recall = (TP_train/(TP_train+FN_train))
    train_precision = (TP_train/(TP_train+FP_train))
    train_f1_score = 2*train_precision*train_recall/(train_precision+train_recall)
    
    lab_train_precision_list.append(train_precision)
    lab_train_recall_list.append(train_recall)
    lab_train_F1_score_list.append(train_f1_score)
    
    TN_test = lab_test_confusion_matrix[0,0]
    FP_test = lab_test_confusion_matrix[0,1]
    FN_test = lab_test_confusion_matrix[1,0]
    TP_test = lab_test_confusion_matrix[1,1]
    test_recall = (TP_test/(TP_test+FN_test))
    test_precision = (TP_test/(TP_test+FP_test))
    test_f1_score = 2*test_precision*test_recall/(test_precision+test_recall)
    
    lab_test_precision_list.append(test_precision)
    lab_test_recall_list.append(test_recall)
    lab_test_F1_score_list.append(test_f1_score)
    
    lab_train_AUC = roc_auc_score(new_labeled_y_train, pd.DataFrame(lab_y_train_predict))
    lab_test_AUC = roc_auc_score(y_test, lab_y_test_predict)
    lab_train_AUC_list.append(lab_train_AUC)
    lab_test_AUC_list.append(lab_test_AUC)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

km_train_accuracy_list = []
km_train_precision_list = []
km_train_recall_list = []
km_train_F1_score_list = []
km_train_AUC_list=[]

km_test_accuracy_list = []
km_test_precision_list = []
km_test_recall_list = []
km_test_F1_score_list = []
km_test_AUC_list=[]

for i in range(30):
    kmeans = KMeans(n_clusters= 2, random_state=i).fit(x_train)
    labels = kmeans.labels_
    #print(labels)
    cluster_centers = kmeans.cluster_centers_
    #print(f"cluster_centers",cluster_centers)
    cluster_centers_0 = cluster_centers[0]
    #print(f"cluster_centers_0", cluster_centers_0)
    cluster_centers_1 = cluster_centers[1]

    train_cluster_0 = np.where(labels==0)
    train_cluster_1 = np.where(labels==1)
    #print(f"train_cluster_0", train_cluster_0)
    
    x_train_cluster_0 = np.array(x_train)[train_cluster_0]
    y_train_cluster_0 = np.array(y_train)[train_cluster_0]

    sum1 = 0
    list1 = []
    k = -1
    for i in x_train_cluster_0:
        k = k + 1
        sum1 = 0
        for j in range(len(i)):
            sub = abs(i[j] - cluster_centers_0[j])
            sum1 = sum1+ sub
        tuple1 =(sum1, k)
        list1.append(tuple1)
    new =sorted(list1)[0:30]

    cluster_list = []
    for i in new:
        cluster_list.append(i[1])
    new_lable = y_train_cluster_0[cluster_list]
    new_lable

    y_train_pred = labels
    y_train_pred[new_lable] = new_lable
    
    x_train_cluster_1 = np.array(x_train)[train_cluster_1]
    y_train_cluster_1 = np.array(y_train)[train_cluster_1]

    sum2 = 0
    list2 = []
    k = -1
    for i in x_train_cluster_1:
        k = k + 1
        sum2 = 0
        for j in range(len(i)):
            sub1 = abs(i[j] - cluster_centers_1[j])
            sum2 = sum1+ sub1
        tuple2 =(sum2, k)
        list2.append(tuple2)
    new2 =sorted(list2)[0:30]

    cluster_list2 = []
    for i in new2:
        cluster_list2.append(i[1])
    new_lable2 = y_train_cluster_1[cluster_list2]

    y_test_pred = kmeans.predict(x_test)
    y_test_pred[new_lable2] = new_lable2
    
    km_train_accuracy = accuracy_score(y_train, y_train_pred)
    km_train_accuracy_list.append(km_train_accuracy)
    km_test_accuracy = accuracy_score(y_test, y_test_pred)
    km_test_accuracy_list.append(km_test_accuracy)
    
    km_train_confusion_matrix = confusion_matrix(y_train, y_train_pred)
    km_test_confusion_matrix = confusion_matrix(y_test, y_test_pred)
    
    TN_train = km_train_confusion_matrix[0,0]
    FP_train = km_train_confusion_matrix[0,1]
    FN_train = km_train_confusion_matrix[1,0]
    TP_train = km_train_confusion_matrix[1,1]
    train_recall = (TP_train/(TP_train+FN_train))
    train_precision = (TP_train/(TP_train+FP_train))
    train_f1_score = 2*train_precision*train_recall/(train_precision+train_recall)
    
    km_train_precision_list.append(train_precision)
    km_train_recall_list.append(train_recall)
    km_train_F1_score_list.append(train_f1_score)
    
    TN_test = km_test_confusion_matrix[0,0]
    FP_test = km_test_confusion_matrix[0,1]
    FN_test = km_test_confusion_matrix[1,0]
    TP_test = km_test_confusion_matrix[1,1]
    test_recall = (TP_test/(TP_test+FN_test))
    test_precision = (TP_test/(TP_test+FP_test))
    test_f1_score = 2*test_precision*test_recall/(test_precision+test_recall)
    
    km_test_precision_list.append(test_precision)
    km_test_recall_list.append(test_recall)
    km_test_F1_score_list.append(test_f1_score)
    
    km_train_AUC = roc_auc_score(y_train, y_train_pred)
    km_test_AUC = roc_auc_score(y_test, y_test_pred)
    km_train_AUC_list.append(km_train_AUC)
    km_test_AUC_list.append(km_test_AUC)
```

```python
from sklearn.cluster import SpectralClustering

spe_train_accuracy_list = []
spe_train_precision_list = []
spe_train_recall_list = []
spe_train_F1_score_list = []
spe_train_AUC_list=[]

spe_test_accuracy_list = []
spe_test_precision_list = []
spe_test_recall_list = []
spe_test_F1_score_list = []
spe_test_AUC_list=[]

for i in range(30):
    clustering1 = SpectralClustering(n_clusters=2,random_state=i)
    clustering1.fit(stand_x_train)

    clustering2 = SpectralClustering(n_clusters=2,random_state=i)
    clustering2.fit(stand_x_test)

    spe_y_train_pred = clustering1.fit_predict(stand_x_train)
    spe_y_test_pred = clustering2.fit_predict(stand_x_test)
    
    spe_train_accuracy = accuracy_score(y_train, spe_y_train_pred)
    spe_train_accuracy_list.append(spe_train_accuracy)
    spe_test_accuracy = accuracy_score(y_test, spe_y_test_pred)
    spe_test_accuracy_list.append(spe_test_accuracy)
    
    spe_train_confusion_matrix = confusion_matrix(y_train, spe_y_train_pred)
    spe_test_confusion_matrix = confusion_matrix(y_test, spe_y_test_pred)
    
    TN_train = spe_train_confusion_matrix[0,0]
    FP_train = spe_train_confusion_matrix[0,1]
    FN_train = spe_train_confusion_matrix[1,0]
    TP_train = spe_train_confusion_matrix[1,1]
    train_recall = (TP_train/(TP_train+FN_train))
    train_precision = (TP_train/(TP_train+FP_train))
    train_f1_score = 2*train_precision*train_recall/(train_precision+train_recall)
    
    spe_train_precision_list.append(train_precision)
    spe_train_recall_list.append(train_recall)
    spe_train_F1_score_list.append(train_f1_score)
    
    TN_test = spe_test_confusion_matrix[0,0]
    FP_test = spe_test_confusion_matrix[0,1]
    FN_test = spe_test_confusion_matrix[1,0]
    TP_test = spe_test_confusion_matrix[1,1]
    test_recall = (TP_test/(TP_test+FN_test))
    test_precision = (TP_test/(TP_test+FP_test))
    test_f1_score = 2*test_precision*test_recall/(test_precision+test_recall)
    
    spe_test_precision_list.append(test_precision)
    spe_test_recall_list.append(test_recall)
    spe_test_F1_score_list.append(test_f1_score)
    
    spe_train_AUC = roc_auc_score(y_train, spe_y_train_pred)
    spe_test_AUC = roc_auc_score(y_test, spe_y_test_pred)
    spe_train_AUC_list.append(spe_train_AUC)
    spe_test_AUC_list.append(spe_test_AUC)
```
