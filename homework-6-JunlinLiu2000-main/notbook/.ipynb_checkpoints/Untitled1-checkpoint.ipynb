```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('../Data/data_banknote_authentication.txt', names =['variance', 'skewness', 'curtosis', 'entropy', 'class'], header = None)
data

x_data = data.iloc[:,0:4]
y_data = data.iloc[:,4]
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=472/1372)
x_train
```

```python
import random
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

test_error_list =[]
c_list = []
for i in range(-3,3):
    c_list.append(10**i)
parameters = {'C':c_list}

x_train1 = x_train
y_train1 = y_train
randomlist = random.sample(range(len(x_train)), 10)
x_train_pool = x_train.iloc[randomlist]
y_train_pool = y_train.iloc[randomlist]
pool_svc = LinearSVC(penalty='l1', dual=False, random_state=i)
pool_clf_train = GridSearchCV(pool_svc, parameters, cv=5).fit(x_train_pool, y_train_pool)
pool_C_train = pool_clf_train.best_params_['C']
pool_svm_train = LinearSVC(penalty='l1', C = pool_C_train, dual=False, random_state = i)
pool_svm_train.fit(x_train_pool, y_train_pool)
y_pool_predict = pool_svm_train.predict(x_test)
test_error = 1 - accuracy_score(y_test, y_pool_predict)
test_error_list.append(test_error)
x_train1.drop(x_train1.index[randomlist])
y_train1.drop(y_train1.index[randomlist])

while len(x_train1) > 0:
    randomlist = random.sample(range(len(x_train1)), 10)
    x_train_pool = x_train_pool.append(x_train1.iloc[randomlist])
    y_train_pool = y_train_pool.append(y_train1.iloc[randomlist])
    pool_svc = LinearSVC(penalty='l1', dual=False, random_state=i)
    pool_clf_train = GridSearchCV(pool_svc, parameters, cv=5).fit(x_train_pool, y_train_pool)
    pool_C_train = pool_clf_train.best_params_['C']
    pool_svm_train = LinearSVC(penalty='l1', C = pool_C_train, dual=False, random_state = i)
    pool_svm_train.fit(x_train_pool, y_train_pool)
    y_pool_predict = pool_svm_train.predict(x_test)
    test_error = 1 - accuracy_score(y_test, y_pool_predict)
    test_error_list.append(test_error)
    x_train1.drop(x_train1.index[randomlist])
    y_train1.drop(y_train1.index[randomlist])
print(test_error_list)
```

```python
import random
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')



c_list = []
for i in range(-3,3):
    c_list.append(10**i)
parameters = {'C':c_list}

test_error_active_list =[]
x_train_active = x_train
y_train_active = y_train
randomlist = random.sample(range(len(x_train)), 10)
x_train_pool_active = x_train_active.iloc[randomlist]
y_train_pool_active = y_train_active.iloc[randomlist]

pool_svc_active = LinearSVC(penalty='l1', dual=False)
pool_clf_train_active = GridSearchCV(pool_svc, parameters, cv=5).fit(x_train_pool_active, y_train_pool_active)
pool_C_train_active = pool_clf_train_active.best_params_['C']
pool_svm_train_active = LinearSVC(penalty='l1', C = pool_C_train_active, dual=False)
pool_svm_train_active.fit(x_train_pool_active, y_train_pool_active)
y_pool_predict_active = pool_svm_train_active.predict(x_test)
test_error_active = 1 - accuracy_score(y_test, y_pool_predict_active)
test_error_active_list.append(test_error_active)
x_train_active.drop(x_train_active.index[randomlist])
y_train_active.drop(y_train_active.index[randomlist])

while len(x_train_active) > 0:
    distance = np.abs(pool_svm_train_active.decision_function(x_train_active))
    tuple_list =[]
    closest_10_point = []
    for i in range(len(x_train_active)):
        tuple_list.append((distance[i],i))
    closest_10 =sorted(tuple_list)[0:10]
    for i in closest_10:
        closest_10_point.append(i[1])
        
    x_train_pool_active = x_train_pool_active.append(x_train_active.iloc[closest_10_point])
    y_train_pool_active = y_train_pool_active.append(y_train_active.iloc[closest_10_point])
    
    pool_svc_active = LinearSVC(penalty='l1', dual=False)
    pool_clf_train_active = GridSearchCV(pool_svc, parameters, cv=5).fit(x_train_pool_active, y_train_pool_active)
    pool_C_train_active = pool_clf_train_active.best_params_['C']
    pool_svm_train_active = LinearSVC(penalty='l1', C = pool_C_train_active, dual=False)
    pool_svm_train_active.fit(x_train_pool_active, y_train_pool_active)
    y_pool_predict_active = pool_svm_train_active.predict(x_test)
    test_error_active = 1 - accuracy_score(y_test, y_pool_predict_active)
    test_error_active_list.append(test_error_active)
    x_train_active.drop(x_train_active.index[closest_10_point])
    y_train_active.drop(y_train_active.index[closest_10_point])

test_error_active
```
