```python
import pandas as pd
import numpy as np

training_data = pd.read_csv('drive/MyDrive/DSCI_552/hw4/Data/aps_failure_training_set.csv', skiprows = 20)
training_data

test_data = pd.read_csv('drive/MyDrive/DSCI_552/hw4/Data/aps_failure_test_set.csv', skiprows = 20)
test_data
```

```python
from sklearn.impute import SimpleImputer
x_training_data = training_data.drop(columns = "class")
y_training_data = training_data[['class']]
x_training_data = x_training_data.replace('na', np.nan)

imp = SimpleImputer(missing_values = np.nan, strategy='mean')
imp.fit_transform(x_training_data)
x_training_data = pd.DataFrame(imp.fit_transform(x_training_data), columns=x_training_data.columns)
x_training_data
```

```python
x_test_data = test_data.drop(columns = "class")
y_test_data = test_data[['class']]
x_test_data = x_test_data.replace('na', np.nan)

imp.transform(x_training_data)
x_test_data = pd.DataFrame(imp.transform(x_test_data), columns=x_test_data.columns)
x_test_data
```

```python
CV = {}
for column in x_training_data.columns:
    std = np.std(x_training_data[column])
    mean = np.mean(x_training_data[column])
    cv = std/mean
    CV[column] = cv
CV
```

## (c)

```python
y_training_data1 = y_training_data
y_training_data1 = y_training_data1.replace('neg', 0)
y_training_data1 = y_training_data1.replace('pos', 1)
y_test_data1 = y_test_data
y_test_data1 = y_test_data1.replace('neg', 0)
y_test_data1 = y_test_data1.replace('pos', 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc

clf = RandomForestClassifier(max_depth = 2, random_state=0, bootstrap = True, oob_score = True)
clf.fit(x_training_data, y_training_data1)
train_pred_y = clf.predict(x_training_data)

yscore = clf.predict_proba(x_training_data)[:,1]
fpr = dict()
tpr = dict()
roc_auc = dict()
fpr, tpr, _ = roc_curve(y_training_data1, yscore)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()

confusionMatrix = confusion_matrix(y_training_data1, train_pred_y)
print(f"confusionMatrix:\n", confusionMatrix)
misclassification = 1 - clf.score(x_training_data, y_training_data1)
print(f"misclassification", misclassification)
oob_error = 1 - clf.oob_score_
print(f'oob_error:', oob_error)
```

```python
clf1 = RandomForestClassifier(max_depth = 2, random_state=0, bootstrap = True, oob_score = True)
clf1.fit(x_test_data, y_test_data1)
train_pred_y = clf1.predict(x_test_data)
test_pred_y = clf1.predict(x_test_data)

yscore1 = clf1.predict_proba(x_test_data)[:,1]
fpr = dict()
tpr = dict()
roc_auc = dict()
fpr, tpr, _ = roc_curve(y_test_data1, yscore1)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()

confusionMatrix = confusion_matrix(y_test_data1, test_pred_y)
print(f"confusionMatrix:\n", confusionMatrix)
misclassification = 1 - clf.score(x_test_data, y_test_data1)
print(f"misclassification", misclassification)
oob_error = 1 - clf.oob_score_
print(f'oob_error:', oob_error)
```

both training data's and test data's oob_error is smaller than test error

```python
import weka.core.jvm as jvm
import weka.core.converters as converters
from weka.classifiers import Classifier
from weka.classifiers import Evaluation
from weka.core.classes import Random
from weka.classifiers import FilteredClassifier
from weka.filters import Filter

new_train = x_training_data
new_train['class'] = y_training_data

new_train.to_csv('weka_train_data.csv', index = False)

jvm.start()

train_data = converters.load_any_file("weka_train_data.csv")
train_data.class_is_last()

remove = Filter(classname="weka.filters.unsupervised.attribute.Remove", options=["-R", "1-3"])

cls = Classifier(classname="weka.classifiers.bayes.NaiveBayes")

fc = FilteredClassifier()
fc.filter = remove
fc.classifier = cls

cls = Classifier(classname="weka.classifiers.trees.LMT")
evl = Evaluation(train_data)
evl.crossvalidate_model(cls, train_data, 5, Random(43))

print(evl.summary())
print(evl.class_details())
```

```python
import weka.plot.classifiers as plcls
plcls.plot_roc(evl, class_index=[0, 1], wait=True)
evl.confusion_matrix
```
