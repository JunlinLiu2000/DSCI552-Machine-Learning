```python
from sklearn.preprocessing import scale 
from sklearn import model_selection
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

mse =[]
pca = PCA()
training_data_x = np.array(x_training_data)
training_data_y = np.array(y_training_data)
x_reduced = pca.fit_transform(scale(training_data_x))
n = len(x_reduced)
    
CV = model_selection.KFold(n_splits=5, shuffle=True, random_state=1)
regr = LinearRegression()
score = -1*model_selection.cross_val_score(regr, np.ones((n,1)), training_data_y.ravel(), cv=CV, scoring='neg_mean_squared_error').mean()    
mse.append(score)

for i in np.arange(1, n):
    score = -1*model_selection.cross_val_score(regr, x_reduced[:,:i], training_data_y.ravel(), cv=CV, scoring='neg_mean_squared_error').mean()
    mse.append(score)

best_mse = min(mse)
best_number = mse.index(best_mse)
print(f'the best number of principal components is:', best_number)
```

```python
test_data_x = np.array(x_test_data)
test_data_y = np.array(y_test_data)
x_reduced_test = pca.fit_transform(scale(test_data_x))[:,:best_number + 1]
regr.fit(x_reduced[:,:best_number + 1], training_data_y)

pred = regr.predict(x_reduced_test)
test_error = mean_squared_error(test_data_y, pred)
print(f'the test error is', test_error)
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(data = x_training_data, label= y_training_data)
alphas = [0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
dic ={}

for a in alphas:
    param = {'max_depth': 2, 'eta': 0.3 , 'silent':1, 'objective': 'reg:linear', 'alpha': a}
    score = xgb.cv(param, dtrain, nfold = 5, metrics = 'rmse')
    mse = np.mean(score['test-rmse-mean'])
    dic[a] = mse
dic 
```

```python
new_dic = sorted((value, key) for (key,value) in dic.items())
best_error = new_dic[0]
print(f'the best α is', best_error[1])
print(f'the mse is', best_error[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = [3, 2, 4, 1, 2, 4, 4]
x2 = [4, 2, 4, 4, 1, 3, 1]
colors = ["red", "red", "red", "red", "blue", "blue", "blue"]
sns.scatterplot(x1, x2, hue = colors)

def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    # from https://stackoverflow.com/questions/7941226/how-to-add-line-based-on-slope-and-intercept-in-matplotlib
abline(1,-0.5)
```
